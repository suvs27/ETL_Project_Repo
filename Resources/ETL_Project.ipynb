```python
# Import the Pandas library
# Importing the CSV Data for June
import pandas as pd
import numpy as np

# Create a reference the CSV file desired
csv_path = "NNDSS_-_TABLE_1V._Malaria_to_Measles__Imported.csv"

# Read the CSV into a Pandas DataFrame
table = pd.read_csv(csv_path)

table.columns

```

```python
#Creating a table related to only measles data

#Dropping information related to malaria
measles = table.drop(columns=
                     ['Malaria, Current week',
                      'Malaria, Current week, flag', 
                      'Malaria, Previous 52 weeks Max†',
                      'Malaria, Previous 52 weeks Max†, flag', 
                      'Malaria, Cum 2019†',
                      'Malaria, Cum 2019†, flag', 
                      'Malaria, Cum 2018†',
                      'Malaria, Cum 2018†, flag', 
                      'Measles§, Idigenousis, Current week, flag',
                      'Measles§, Idigenousis, Cum 2019†, flag',
                      'Measles§, Idigenousis, Previous 52 weeks Max†, flag',
                      'Measles§, Idigenousis, Cum 2019†, flag',
                      'Measles§, Idigenousis, Cum 2018†, flag',
                      'Measles§, Imported, Current week, flag',
                      'Measles§, Imported, Previous 52 weeks Max†, flag',
                      'Measles§, Imported, Cum 2019†, flag',
                      'Measles§, Imported, Cum 2018†, flag',
                      'Location 1',
                      'Location 2'])

#Filling in NAN values with UNAVAILABLE
#measles
#measles = measles.fillna("UNAVAILABLE")


# Giving columns simpler names to facilitate querying
measles = measles.rename(columns={"MMWR Year": "Year", 
                        "MMWR Week": "Week", 
                        "Measles§, Idigenousis, Current week": "Measles Indigenousis Current Week",
                        "Measles§, Idigenousis, Previous 52 weeks Max†": "Measles Indigenousis Previous 52 Weeks Max",
                        "Measles§, Idigenousis, Cum 2019†":"Measles Indigenousis Cum 2019",
                        "Measles§, Idigenousis, Cum 2018†": "Measles Indigenousis Cum 2018",
                        "Measles§, Imported, Current week":"Measles Imported Current Week",
                        "Measles§, Imported, Previous 52 weeks Max†":"Measles Imported Previous 52 Weeks Max",
                        "Measles§, Imported, Cum 2019†":"Measles Imported Cum 2019",
                        "Measles§, Imported, Cum 2018†":"Measles Imported Cum 2018"})

measles


```

```python
# Creating a table to compare Max Count of Reported Measles cases in 2019 for Indigenousis v. Imported
measles_max = measles[["Week", "Reporting Area", "Year", "Measles Indigenousis Previous 52 Weeks Max", "Measles Imported Previous 52 Weeks Max"]]
measles_max

#Check number of nan values
#Here count() tells us the number of non-NaN values
#and this is subtracted from the total number of values (given by len(df))
nan_num = len(measles_max) - measles_max.count()
nan_num

measles_max = measles_max.dropna(how='any')
measles_max = measles_max.reset_index(drop=True)
measles_max

measles_max_current = measles_max[measles_max["Week"]==30]
measles_max_current
measles_max_current = measles_max_current.sort_values("Reporting Area")
measles_max_current = measles_max_current.reset_index(drop=True)
measles_max_current["Combined Preveious 52 Weeks Max"] = measles_max_current["Measles Indigenousis Previous 52 Weeks Max"] + measles_max_current["Measles Imported Previous 52 Weeks Max"]
measles_max_current = measles_max_current.sort_values("Combined Preveious 52 Weeks Max", ascending=False).reset_index(drop=True)
#This table has the latest information for max measles indigenousis and max measles imported for the reporting areas up to Week 30
measles_max_current






```

```python
measles_cumulative = jsondf[['Reporting Area',
                             'MMWR Year',
                             'MMWR Week',
                             'MeaslesÂ§, Idigenousis, Cum 2019â€\xa0',
                            'MeaslesÂ§, Idigenousis, Cum 2018â€\xa0',
                            'MeaslesÂ§, Imported, Cum 2019â€\xa0',
                            'MeaslesÂ§, Imported, Cum 2018â€\xa0']]

measles_cumulative

measles_cumulative = measles_cumulative.rename(columns={
                        "MMWR Year": "Year", 
                        "MMWR Week": "Week", 
                        'MeaslesÂ§, Idigenousis, Cum 2019â€\xa0':"Measles Indigenousis Cum 2019",
                        'MeaslesÂ§, Idigenousis, Cum 2018â€\xa0': "Measles Indigenousis Cum 2018",
                        'MeaslesÂ§, Imported, Cum 2019â€\xa0': "Measles Imported Cum 2019",
                        "Measles§, Imported, Cum 2018†":"Measles Imported Cum 2018",
                        'MeaslesÂ§, Imported, Cum 2018â€\xa0': "Measles Imported Cum 2018"})

measles_cumulative = measles_cumulative.dropna(how="any")
measles_cumulative

measles_cumulative = measles_cumulative.astype({"Measles Indigenousis Cum 2019": 'int', 
                           "Measles Imported Cum 2019":'int',
                          "Measles Indigenousis Cum 2018": 'int',
                          "Measles Imported Cum 2018": 'int',})



measles_cumulative["Measles 2019 Combined"] = measles_cumulative["Measles Indigenousis Cum 2019"] + measles_cumulative["Measles Imported Cum 2019"]
measles_cumulative["Measles 2018 Combined"] = measles_cumulative["Measles Indigenousis Cum 2018"] + measles_cumulative["Measles Imported Cum 2018"]
measles_cumulative.reset_index()
measles_cumulative 



grp_measles_cumulative = measles_cumulative.groupby(["Reporting Area"])
total2 = grp_measles_cumulative["Measles 2019 Combined"].sum()


measles_combined_available = pd.DataFrame({"Combined Measles Cases in 2019":total2}).sort_values("Combined Measles Cases in 2019")

measles_combined_available


```

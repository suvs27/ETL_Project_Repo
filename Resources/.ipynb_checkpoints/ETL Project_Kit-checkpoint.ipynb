```python
# URL of page to be scraped
url = 'https://www.cdc.gov/mmwr/volumes/68/wr/mm6817e1.htm'

# Retrieve page with the requests module
response = requests.get(url)
# Create BeautifulSoup object; parse with 'lxml'
soup = bs(response.text, 'lxml')
```

```python
soup.find_all('table')
```

```python
cdc_table = pd.read_html(url)[0]
#age_group_tbl = age_group_tbl.rename(columns={"Characteristic":"Age Group"})
cdc_table = cdc_table.rename(columns={"No. (%)":"Number(%)"})
cdc_table
```

```python
age_group_tbl = cdc_table[1:]
age_group_tbl = age_group_tbl[:-25]
age_group_tbl = age_group_tbl.rename(columns={"Characteristic":"Age Group"})
age_group_tbl = age_group_tbl[1:]
age_group_tbl = age_group_tbl.reset_index(drop=True)
age_group_tbl
```

```python
vaccination_tbl = cdc_table[1:]
vaccination_tbl = vaccination_tbl[:-21]
vaccination_tbl = vaccination_tbl.rename(columns={"Characteristic":"Vaccination Status"})
vaccination_tbl = vaccination_tbl[9:]
vaccination_tbl = vaccination_tbl.reset_index(drop=True)
vaccination_tbl
```

```python
hospitalizations_tbl = cdc_table[13:]
hospitalizations_tbl = hospitalizations_tbl[:-20]
hospitalizations_tbl = hospitalizations_tbl.rename(columns={"Characteristic":"Hospitalizations"})
hospitalizations_tbl = hospitalizations_tbl.reset_index(drop=True)
hospitalizations_tbl
```

```python
residency_tbl = cdc_table[19:]
residency_tbl = residency_tbl[:-14]
residency_tbl = residency_tbl.rename(columns={"Characteristic":"Residency"})
residency_tbl = residency_tbl.reset_index(drop=True)
residency_tbl
```
